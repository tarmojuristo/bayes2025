# Project

The data was gathered by general practitioners during patients' visits with no history of liver cirrhosis or liver transplant. Every patient was measured, and their waist-to-hip ratio (WHR) was calculated. Patients' physical activity and alcohol consumption was self-reported and alcohol consumption was converted to weekly consumption units. Also, sex and age of each patient were recorded. Patients aged 50 to 70 years were included into the sample. After data collection, their future outcomes of liver cirrhosis and liver transplant were observed during a 10-year period.

### Aim
We are interested in the average total effect of weekly alcohol consumption on cirrhosis.

### Dataset (n=1000)
- **age** - patient's age (numerical variable)
- **sex** - patient's sex (categorical variable: "Male"/"Female")
- **physical_activity** - patient's physical activity level (categorical variable: "Low"/"Moderate"/"High")
- **alcohol_units** - alcohol consumption (numerical variable; units per week) 
- **whr** - patient's waist-to-hip ratio (numerical variable)
- **cirrhosis** - binary variable indicating whether the patient was diagnosed with liver cirrhosis within 10 years (0 = no, 1 = yes)
- **liver_transplant** - binary variable indicating whether the patient received a liver transplant within 10 years (0 = no, 1 = yes)
